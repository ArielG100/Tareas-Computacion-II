- Agregar pasajeros a la lista de viajeros.

- Dado el DNI de un pasajero, ver a qué ciudad viaja.  
- Dada una ciudad, mostrar la cantidad de pasajeros que viajan a esa ciudad.  
- Dado el DNI de un pasajero, ver a qué país viaja.  
- Dado un país, mostrar cuántos pasajeros viajan a ese país.

```python

print('Transporte Aereo Paraguayo')
list1 = [("Manuel Juarez", 19823451, "Liverpool"), ("Silvana Paredes", 22709128, "Buenos Aires"),
         ("Rosa Ortiz", 15123978, "Glasgow"), ("Luciana Hernandez", 38981374, "Lisboa"),('Ariel García', 2429917, 'Liverpool')]
list2 = [("Buenos Aires","Argentina"), ("Glasgow","Escocia"), ("Lisboa", "Portugal"),
         ("Liverpool","Inglaterra"), ("Madrid","España"), ('Oporto', 'Portugal')]


def Agregar_Pasajeros():
    pasajero =[]
    pasajero.append(input('Nombre: '))
    pasajero.append(int(input('DNI: ')))
    pasajero.append(input('Destino: '))
    list1.append(tuple(pasajero))
    print (list1)

def Agregar_Cuidades():
    city = []
    city.append(input('Ciudad: '))
    city.append(input('País: '))
    list2.append(tuple(city))
    print(list2)
    
def Ver_Destino_Ciudad(lista=list1):
    dni = int(input('Ingresa DNI: '))
    for l in lista:
        if(l[1]==dni):
            print('El destino para',l[0],'es',l[2])

    
def Pasajeros_por_Ciudad():
    newlist=[]
    city = input('Ciudad destino: ')
    for l in list1:
        newlist.append(l[2])    #se crea lista de ciudades viajadas
    print('Viajan',newlist.count(city),'personas a',city) #se cuentan las ocurrencias
    
#En esta funcion reemplazo las ciudades por paises de list1 convirtiendo
#antes las tuplas en listas para hacerlas mutables, luego con la nueva
#lista que contiene paises asociados a los DNI utilizo la funcion
#previamente creada 'Ver_Destino_Ciudad(listpais)' que recibe la nueva lista de argumento
def Ver_Destino_Pais(): 
    listalista = []        
    for l in list1:
        tup = list(l)    #Convierto las tuplas de list1 en listas
        listalista.append(tup)  #Se crea una lista de listas
    diccio = dict(list2)    #Convierto la list2 en dicionario
    listpais = []
    for l in listalista:
        l[2] = diccio[l[2]] #Reemplazo ciudad por pais
        listpais.append(l)  #se crea lista nueva con pais en lugar de ciudad
    Ver_Destino_Ciudad(listpais)    #ejecutar funcion anterior con nuevo argumento

def Pasajeros_por_Pais():
    newlist=[]
    diccio = dict(list2)    #Se crea un dicc para usar etiquetas-valores en el bucle for
    pais = input('Pais destino: ')
    for l in list1:
        newlist.append(diccio[l[2]])    #se crea lista de paises viajados en lugar de ciudades
    print('Viajan',newlist.count(pais),'personas a',pais)   

def error():
    print('No Válido')

switch_opcion = {
	1: 'Agregar_Pasajeros',
    2: 'Agregar_Cuidades',
	3: 'Ver_Destino_Ciudad',
    4: 'Pasajeros_por_Ciudad',
    5: 'Ver_Destino_Pais',
    6: 'Pasajeros_por_Pais',
    0: 'Salir'
}

while(True):
    print('---------------------------')
    for l in switch_opcion:
        print(l,switch_opcion[l],end=' | ')
    opt = int(input('Ingrese opcion: '))
    if (opt==0):
        break
    else:
        eval(switch_opcion.get(opt, 'error')+'()')

```

- Informar qué nombres se repiten entre los alumnos de nivel primario y secundario.
- Informar qué nombres de nivel primario no se repiten en los de nivel secundario.

```python
print('Alumons de ciclo inicial')
primer =  ['Ari','Clau','Bero','Rubi','Lili','Mike','Fred']
secundario = ['Fred','Josh','Julio','Clau','Ari']
nombre = ''
while (nombre != 'x'):
    nombre = input('Nivel primario. Ingresar nombre: ')
    primer.append(nombre)
    if nombre == 'x': primer.pop(-1)
nombre = ''
while (nombre != 'x'):
    nombre = input('Nivel secundario. Ingresar nombre: ')
    secundario.append(nombre)
    if nombre == 'x': secundario.pop(-1)

todos = primer + secundario    
print('Todos los alumnos del ciclo inicial')
print(set(todos))

rep = []
for l in primer:
    if (l in secundario):
        rep.append(l)
print('Alumnos que se repiten entre el 1° y 2°')
print (rep)

norep=[]
for l in primer:
    if not(l in secundario):
        norep.append(l)
print('Alumnos del 1° que no se repiten en el 2°')
print(norep)
```

```python
lista = []
cadena=''
for n in range(5): #Solo usé 5 en lugar de 50 para escribir 5 frases largas
    texto = input('Ingrese su texto: ')
    print(texto)
    cadena = cadena + texto
print('Buscando todos los caracteres\n----------')
for s in cadena:
    lista.append((s,cadena.count(s)))
print(dict(sorted(lista)))


print('Buscando alfabeto unicamente\n----------')
mayus =[]
minus =[]
for s in range(65,91):  #Valores de las mayusculas para el codigo ASCII
    mayus.append((chr(s),cadena.count(chr(s))))
for s in range(97,123): #Valores de las minusculas 
    minus.append((chr(s),cadena.count(chr(s))))
print(dict(sorted(mayus+minus)))
```

6. Crear un programa para gestionar datos de los socios de un club, permitiendo:
- Cargar información de los socios en un diccionario para acceder por número de socio.  
Los datos a almacenar son: número, nombre y apellido, fecha de ingreso (ddmmaaaa), cuota al día (s/n).  
El programa debe iniciar con los datos de los socios fundadores ya cargados:  
  - Socio nº1, Amanda Núñez, ingresó: 17/03/2009, cuota al día.  
  - Socio nº2, Bárbara Molina, ingresó: 17/03/2009, cuota al día.  
  - Socio nº3, Lautaro Campos, ingresó: 17/03/2009, cuota al día.  
- Informar cantidad de socios del club.
- Solicitar al usuario el número de un socio y registrar que ha pagado todas las cuotas adeudadas.
- Modificar la fecha de ingreso de todos los socios ingresados el 13/03/2018, para indicar que en realidad ingresaron el 14/03/2018.
- Solicitar el nombre y apellido de un socio y darlo de baja (eliminarlo del listado).
- Imprimir el listado de socios completo.

```python
socios = {1:['Amanda Nuñez','17032009','s'],2:['Barbara Molina','15062009','s'],3:['Lautaro Campos','23022009','s']}

def Agregar_Socio():  #El número de socio se carga en forma automática
    num = list(socios.keys())[-1] + 1   #El ultimo numero de socio + 1
    print('Ingrese el nombre y apellido del socio N°',num,end=':')
    nombre = input(' ')
    fecha = input('Fecha de ingreso (ddmmaaaa): ')
    cuota = input('Cuota al día? (s/n)')
    socios[num] = [nombre,fecha,cuota]
    
def Cantidad():
    print('El N° de socios es:',len(socios))
    
def Cuotas_al_Dia():
    numero = int(input('Ingrese numero de Socio: '))
    nombre = socios[numero][0]
    cuota = socios[numero][2]
    print(nombre,'se encuentra al dia con las cuotas (',cuota,'):')
    cuota = input('(s/n): ')
    socios[numero][2] = cuota

def Modificar_Fecha():
    fecha = input('Ingrese fecha a modificar (ddmmaaa): ')
    nuevafecha = input('Ingrese fecha correcta (ddmmaaa): ')
    for k in socios:
        if socios[k][1] == fecha:
            socios[k][1] = nuevafecha
        
def Eliminar_Socio():
    nombre = input('Ingrese nombre y apellido para eliminar: ')
    for k in range(1,len(socios)+1):
        if (socios[k][0]==nombre):
            socios.pop(k)

def Imprimir_Socios():
    for k in socios:
        print('Socio N°{0}, {1}, ingresó en fecha: {2}, cuota'.format(k,socios[k][0],socios[k][1]),end='')
        if (socios[k][2]=='s'):
            print(' al dia')
        else:
            if (socios[k][2]=='n'):
                print('retrasada')
            else:
                print(' Error')
    
switch_opcion = {
	1: 'Agregar_Socio',
    2: 'Cantidad',
	3: 'Cuotas_al_Dia',
    4: 'Modificar_Fecha',
    5: 'Eliminar_Socio',
    6: 'Imprimir_Socios',
    0: 'Salir'
}

while(True):
    print('---------------------------')
    for l in switch_opcion:
        print(l,switch_opcion[l])
    opt = int(input('Ingrese opcion: '))
    if (opt==0):
        break
    else:
        eval(switch_opcion.get(opt, 'error')+'()')
```
