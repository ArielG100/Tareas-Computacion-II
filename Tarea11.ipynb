Ajuste la línea de regresión usando el método de mínimos cuadrados, imprime el valor de la pendiente, la intersección, el coeficiente de correlación. Muestra la gráfica.

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [30,28,32,25,25,25,22,24,35,40]
y = [25,30,27,40,42,40,50,45,30,25]
pendiente, interseccion, r, p, std_err = stats.linregress(x, y)

def mifuncion(x):
#y = ax + b donde a=pendiente y b=interseccion
    return pendiente* x + interseccion
mimodelo = list(map(mifuncion, x))
print('La pendiente es:',pendiente)
print('La interseccion es:',interseccion)
print('El coheficiente de correlación es:',r)
plt.scatter(x, y)
plt.plot(x, mimodelo)
plt.show()
```

2 . Busca un problema de regresión lineal simple de interés y resuelve utilizando python. Explica el paso a paso.

```python
from scipy import stats
from sklearn import datasets    #Librería para python con varias herramientas y una base de datos de ejemplo
import matplotlib.pyplot as plt

boston = datasets.load_boston() #se carga los datos de viviendas disponibles en sklearn
#print(boston.keys())   #Los datos estan en un diccionario. Esto se usa para ver las claves disponibles
#print(boston.feature_names) #Nombre de todas las columnas de datos
#print(boston.DESCR)    #Descripcion de la informacion de cada columna
x = boston.data[:,5]    #tomamos la 5ta columna que es el numero de habitaciones por cada vivienda
y = boston.target       #Tomamos el precio de la vivienda

pendiente, interseccion, r, p, std_err = stats.linregress(x, y) #Obtenemos los datos estadisticos

def mifuncion(x):
#y = ax + b donde a=pendiente y b=interseccion
    return pendiente* x + interseccion
mimodelo = list(map(mifuncion, x))
print('La pendiente es:',pendiente)
print('La interseccion es:',interseccion)
print('El coheficiente de correlación es:',r)
plt.subplots(figsize=(8, 8)) #Cambio el tamaño de la figura de salida
plt.scatter(x, y)   
plt.xlabel('Numero de habitaciones')    #etiqueta de fila
plt.ylabel('Valor medio')       #etiqueta de columna
plt.plot(x, mimodelo,color='red',linewidth=3)
plt.show()
```
